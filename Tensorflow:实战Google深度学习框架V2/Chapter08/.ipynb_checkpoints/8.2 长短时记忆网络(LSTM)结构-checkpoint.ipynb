LSTMLSTM
$$
\begin{aligned}
z &= tanh(W_z[h_{t-1}, x_t])\, () \\
i &= sigmoid(W_i[h_{t-1}, x_t])\, ()\\
f &= sigmoid(W_f[h_{t-1}, x_t])\, ()\\
o &= sigmoid(W_o[h_{t-1}, x_t])\, ()\\
c_t &= f \cdot c_{t-1} + i \cdot z\, ()\\
h_t &= o\cdot tanh c_t\, ()
\end{aligned}
$$
