MNIST手写体识别数据集是一个比较简单的数据集，包含了60000张图片作为训练数据，10000张图片作为测试数据。在MNIST数据集中的每一张图片都代表0~9中的一个数字。图片大小都为28$\times$28，且数字会出现在图片的中间。目前通过使用卷积神经网络达到的错误率已经非常接近人工标记标记的错误。<br/>

另外CIFAR数据集是一个影响力很大的数据集，分为CIFAR-10和CIFAR-100两个问题，它们都是图像词典项目中800万张图片的一个子集。CIFAR数据集中图片的大小为$32\times 32$的彩色图片。
CIFAR-10收集了来自10个不同种类的60000张图片，与MNIST类似，CIFAR-10数据集中每张图片的大小是固定的，并且仅包含一个种类实体。但和MNIST相比，CIFAR-10数据集最大区别图片由黑白变成彩色，分类难度也比较高。在CIFAR-10数据集上，人工标注的正确率大概为94%，这比MNIST数据集上人工表现要低得多。目前通过使用卷积神经网络，目前CIFAR-10最好的图像识别算法正确率是95.59%。<br/>
在现实中图片的分辨率不会是固定大小，种类不止10种或者100种，且每张图片的不会只出现一个物体。为了是图像识别问题更加贴近现实生活，由Feifei Li教授带头整理的ImageNet很大程度解决了这些问题。<br/>
ImageNet是基于WordNet的大型图像数据库，在ImageNet中，将近1500万图片被关联到WordNet的大约20000个名词同义词集上。在ImageNet的图片中，一张图片中可能出现多个同义词集所代表的实体。ImageNet每年都会举办图像识别相关的竞赛（ImageNet Large Scale Visual Recognition Chanllenge，ILSVRC），并且每年的竞赛都会有一些不同的问题，这些问题基本涵盖了图像识别的主要研究方向。不同年份的ImageNet比赛提供了不同的数据集。<br/>
ILSVRC2012图像分类数据集的任务和CIFAR数据集是一致的，也是图像识别的主要物体。ILSVRC2012图像分类数据集包括了来自1000个类别的120万张图片，其中每张图片只属于一个类别，并且每张图片大小不一。<br/>

ImageNet问题中，在复杂问题上，基于卷积神经网络的图像识别算法可以远远超过人类的表现。<br/>
