## LeNet-5模型
LeNet-5模型是Yann LeCun在1998年论文`Gradient-based learning applied to documnet recognition`中提出的。在MNIST数据集上，LeNet-5模型可以达到大约99.2%的正确率。LeNet-5模型总共有7层，LeNet-5模型的架构为：

LeNet-5模型每一层的结构：
- `第一层，卷积层`<br/>
这一层的输入就是原始的图像像素。LeNet-5模型接收的输入层大小为$32\times 32\times 1$，第一层的filter尺寸为$5\times 5$，深度为6，不使用全0填充，步长为1。<br/>
- `第二层，池化层`<br/>
这一层的输入时第一层的输出，是一个$28\times 28\times 6$的节点矩阵。本层采用filter大小为$2\times 2$，长和宽的步长为2，所以本层的输出矩阵是$14\times 14\times 6$。<br/>
- `第三层，卷积层`<br/>
这一层的输入是第二层的输出，使用filter的大小是$5\times 5$，深度为16。本层不使用全0填充，步长为1。本层输出矩阵大小为$10\times 10\times 16$。<br/>
- `第四层，池化层`<br/>
这一层的输入是第三层的输出，采用过滤器大小为$2\times 2$，步长为2。本层的输出矩阵为$5\times 5\times 16$。<br/>
- `第五层，全连接层`<br/>
本层的输入是第四层的输出。<br/>
- `第六层，全连接层`<br/>
本层输入节点个数为120个，输出节点个数为84个。<br/>
- `第7层，全连接层`<br/>
本层输入节点数为84个，输出节点数为10个。

一些用于图片分类的卷积神经网络，可以使用如下正则表达式：
$$
卷积层\rightarrow (卷积层+\rightarrow 池化层?)+\rightarrow 全连接层+
$$
大部分卷积神经网络中一般最多连续使用三个卷积层，池化层虽然可以起到减少参数防止过拟合问题，但是大部分论文中发现可以直接通过调整卷积层步长来完成。在多个卷积层和池化层之后，卷积神经网络一般再输出之前会经过1~2层全连接层。<br/>
除了LeNet-5模型，AlexNet、ZF Net和VGGNet的神经网络架构都满足上述的的正则表达式。<br/>
VGGNet模型的作者在论文`Very Deep Convolutional Networks for Large-Scale Image Recognition`中尝试不同的卷积神经网络架构。

表6-1中，convX-Y表示过滤器的尺寸为X，深度为Y。VGGNet中过滤器的尺寸一般为1或者3，LeNet-5模型的过滤器的尺寸也使用了5。一般神经网络的过滤器尺寸不超过5，但是也有一些使用了7或者11。<br/>
在模型深度上，大部分卷积神经网络采用的逐层递增的方式，比如在表6-1中可以看到，每经过一次池化层之后，卷积层过滤器的深度就乘以2。虽然不同模型会选择不同的具体数字，但是逐层递增是比较普遍的模式。卷积层的步长一般为1，但是有些模型也会使用2或者3作为步长。<br/>
池化层的配置相对简单，用的最多的是最大池化层，池化层的过滤器尺寸一般为2或3，步长一般也为2或3。
