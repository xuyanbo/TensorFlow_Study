神经网络中没两层之间的所有节点都有边相连，这个神经网络结构称之为全连接层神经网络，又称为全连接神经网络。卷积神经网络也是通过层与层之间的节点组织起来，但是相邻两层之间只有部分节点相连。<br/>

除了结构相似之外，卷积神经网络的输入输出和训练过程与全连接神经网络也基本一致。卷积神经网络和全连接神经网络的唯一区别在于神经网络相邻两层的连接方式。<br/>
使用全连接神经网络处理图像的最大问题在于全连接层的参数太多。对于MNIST数据集，图片是$28 \times 28 \times 1$的大小，假设第一个隐藏层节点数是500，那么一个全连接层的神经网络将有$28 \times 28 \times 500 + 500 = 392500$个参数。当图片更大，比如CIFAR-10数据集，图片是$32 \times 32 \times 3$的大小，假设隐藏层节点数为500，那么这一层全连接网络将有$32 \times 32 \times 3 \times 500 + 500 \approx 150$万个参数。<br/>
参数增多除了导致计算速度减慢，还可能导致过拟合的问题，因此卷积神经网络的引入，可以有效的解决此问题。<br/>

一个卷积神经网络主要有5部分组成：<br/>
- `输入层。`输入层是整个神经网络的输入，在处理图像的卷积神经网络中，它一般代表了一张图片的像素矩阵。从输入层开始，卷积神经网络通过不同的神经网络结构将上一层的三维矩阵转化为下一层的三维矩阵，直到最后的全连接层。<br/>
- `卷积层。`卷积层是卷积神经网络重要组成部分，卷积层中每个节点的输入只是上一层神经网络的一小块，这小块通常是$2\times 2$、$3\times 3$和$5\times 5$。卷积层试图将神经网络中的一小块进行更加深入地分析从而得到抽象程度更高的特征。一般来说，通过卷积层处理过的节点矩阵会变得更深。<br/>
- `池化层。`池化层不会改变节点矩阵的深度，而只是缩小矩阵的大小。池化操作可以认为将一张分辨率较高的图片转化为分辨率较低的图片，通过池化层，可以进一步缩小最后全连接层中节点的个数，从而达到减少整个神经网络中参数的目的。<br/>
- `全连接层。`在经过多层的卷积层和池化层操作，图像中的信息被抽象为信息含量更高的特征，最后一般再由1到2个全连接层来给出最后的分类结果。<br/>
- `Softmax层。`Softmax层主要用于分类问题。<br/>
