## MNIST数据处理
MNIST数据集是NIST数据集的一个子集，包含了60000张图片作为训练数据，10000张图片作为测试数据，这个数据集通过作为深度学习的入门样例。这个数据集只提供训练数据和测试数据，为了验证模型训练效果，一般会从训练数据中划出一部分数据作为验证数据。TensorFlow为了方便使用该数据集，提供了一个类用于处理MNIST数据集。这个类会自动下载并转化MNIST数据的格式，将数据从原始的数据包中解析成训练和测试神经网络时使用的格式。<br/>
TensorFlow读取MNIST数据集样例：

```python
from tensorflow.examples.tutorials.mnist import input_data
# 加载数据集，如果目录下没有该数据集则自动下载
mnist = input_data.read_data_sets('../Data', one_hot=True)
# 打印训练数据集大小
print('Training Data Size:', mnist.train.num_examples)
# 打印验证数据集大小
print('Validate Data Size:', mnist.validation.num_examples)
# 打印测试集大小
print('Testing Data Size:', mnist.test.num_examples)
# 打印训练集数据内容
print('Example Training Data:', mnist.train.images[0])
# 打印训练集数据标签
print('Example Trainning Label:', mnist.train.labels[0])
# 为了方便使用随机梯度下降，还提供了读取batch训练集的函数
batch_size = 100
xs, ys = mnist.train.next_batch(batch_size)
print('X shape', xs.shape)
print('Y shape', ys.shape)
```
