梯度下降算法主要用于优化单个参数的取值，而反向传播算法通过对神经网络模型上的所有待优化参数使用梯度下降算法，从而使神经网络在训练数据上的损失函数尽可能小。<br/>
梯度下降算法的思想为：假设用$\theta$表示神经网络中的参数，$J(\theta)$表示在给定的参数取值下，训练数据集上损失函数的大小。那么整个神经网络模型的优化过程可以看成通过寻找一个参数$\theta$，使得$J(\theta)$最小。在目前还没有通用方法来通过对任意损失函数直接求解来获取最佳的参数值，所以在实践中，梯度下降算法是最常用的神经网络优化方法。**梯度下降算法会迭代式更新参数$\theta$，不断沿着梯度的反方向让参数朝着总损失更小的方向更新。**<br/>

参数的梯度可以通过求偏导的方式计算，对于参数$\theta$，其梯度为$\frac{\partial}{\partial \theta}J(\theta)$，同时定义一个学习率$\eta$来指定每次参数更新的幅度。通过参数的梯度和学习率，参数的更新公式为：
$$\theta_{n+1}=\theta_{n}-\eta \frac{\partial}{\partial \theta_{n}}J(\theta)$$
神经网络的优化过程大致可以分为两个阶段：
1. 通过前向传播算法计算得到预测值，并利用定义的损失函数计算预测值和真实值之间的差值；
2. 通过反向传播算法计算损失函数对每个参数的梯度，再根据梯度和学习率使用梯度下降算法更新每一个参数。

梯度下降算法并不能保证被优化的函数得到全局最优解，有可能只能得到局部最优解。可见对于参数初始值的不同可能会较大程度影响最后的结果，只有当损失函数为凸函数时，梯度下降算法才能达到全局最优解。<br/>
除了不一定能达到全局最优解，梯度下降算法的另一个问题是计算时间太长。因为每次迭代都是全部训练数据上的最小化损失，所以损失函数$J(\theta)$是在所有训练数据上的损失和，这导致训练的时候非常消耗时间。<br/>
为了加速训练过程，可以采用随机梯度下降算法(stochastic gradient descent)，该算法是每次随机优化某一条训练数据上的损失函数，而不需要使用全部的训练数据，这大大加快了每一轮参数更新的速度。但是使用随机梯度下降算法得到的神经网络可能无法达到局部最优。<br/>
综合以上两个算法的优缺点，实践中一般采用每次计算一小部分数据的损失函数，这一小部分数据被称为一个batch。通过矩阵计算，每次在一个batch上优化神经网络的参数并不会比单位数据慢太多，另外，每次使用一个batch可以大大减少收敛所需要的迭代次数，同时使收敛效果更加接近梯度下降的效果。
