维基百科对深度学习精确的定义为：<br/>
> 一类通过多层非线性变换对高复杂性数据建模算法的集合。

因为深层神经网络是实现"多层非线性变换"最常见的方式，因此深度学习也是深层神经网络的代名词。根据定义可知，深度学习网络有两个非常重要特性：非线性和多层。<br/>
## 线性模型的局限性
在线性模型中，模型的输出是所有输入的加权和。假设一个模型的输出$y$和输入$x_i$满足以下关系：<br/>
$$y = \sum_{i}w_ix_i + b$$
其中$w_i, b \in R$是模型的参数。<br/>
线性模型的最大特点是任意线性模型的组合仍然还是线性模型。因此在线性模型下，多层神经网络任意层的全连接神经网络和单层神经网络模型表达能力没有任何区别。然而线性模型是神经网络的激活函数为线性函数，在平面上表现多直线或者超平面，因此只能解决简单的问题，无法解决复杂的问题。因此对激活函数使用非线性函数，来达到解决复杂的问题。<br/>
## 激活函数实现去线性化
如果对神经元的输出为所有输入加权和，这将导致神经网络模型为线性模型。如果将所有神经元的输出通过一个非线性函数，那么就能得到非线性的神经网络模型，这个非线性函数称为`激活函数`。<br/>
